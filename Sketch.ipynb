```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show(title, img, cmap = "RdBu"):
    plt.imshow(img, cmap = cmap)
    plt.title(title)
```

```python
image = plt.imread("Rose.jpg")
show("Rose",image)
```

```python
Grey_Sketch = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
show("Sketch",Grey_Sketch, 'gray' )
```
